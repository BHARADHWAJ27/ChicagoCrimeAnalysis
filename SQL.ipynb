```python
Q1 = """
SELECT `Primary_Type`, COUNT(*) AS Total_Crimes
FROM CCA.crime_data
GROUP BY `Primary_Type`
ORDER BY Total_Crimes DESC;
"""

cursor = connection.cursor()
cursor.execute(Q1)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

```python
q2="""
SELECT
    HOUR(Time) AS crime_hour,
    COUNT(*) AS total_crimes
FROM CCA.crime_data
GROUP BY crime_hour
ORDER BY total_crimes DESC;
"""
cursor = connection.cursor()
cursor.execute(q2)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

```python
q3="""
SELECT
Year,
COUNT(*) AS total_crimes
FROM CCA.crime_data
Group By Year
ORDER BY Year;
"""
cursor = connection.cursor()
cursor.execute(q3)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

```python
q4 = """
SELECT
    `Primary_Type`,
    COUNT(*) AS total_crimes,
    ROUND((COUNT(*) / (SELECT COUNT(*) FROM crime_data)) * 100, 2) AS percentage
FROM CCA.crime_data
GROUP BY `Primary_Type`
ORDER BY percentage DESC;
"""

cursor = connection.cursor()
cursor.execute(q4)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

```python
q5 = """
SELECT
    `Primary_Type`,
    COUNT(*) AS total_crimes,
    HOUR(STR_TO_DATE(`Time`, '%H:%i:%s')) AS crime_hour
FROM CCA.crime_data
GROUP BY `Primary_Type`, crime_hour
ORDER BY `Primary_Type`, total_crimes DESC;
"""
cursor = connection.cursor()
cursor.execute(q5)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

```python
q6 = """
SELECT District,`Primary_Type`,COUNT(*) AS total_crimes
FROM CCA.crime_data
GROUP BY District,`Primary_Type`
HAVING total_crimes<5
ORDER BY District, total_crimes;
"""
cursor = connection.cursor()
cursor.execute(q6)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

```python
q7 = """
SELECT `Primary_Type`,
SUM(arrest) AS total_arrests,
COUNT(*) AS total_crimes,
ROUND(SUM(arrest) / COUNT(*) * 100, 2) AS arrest_rate
FROM CCA.Crime_Data
GROUP BY `Primary_Type`
ORDER BY arrest_rate DESC;

"""
cursor = connection.cursor()
cursor.execute(q7)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

```python
q8 = """
SELECT
    CASE
        WHEN MONTH(Case_Date) IN (12, 1, 2) THEN 'Winter'
        WHEN MONTH(Case_Date) IN (3, 4, 5) THEN 'Spring'
        WHEN MONTH(Case_Date) IN (6, 7, 8) THEN 'Summer'
        ELSE 'Autumn'
    END AS season,
    COUNT(*) AS total_crimes
FROM CCA.crime_data
GROUP BY season
ORDER BY FIELD(season, 'Winter', 'Spring', 'Summer', 'Autumn');
"""

cursor = connection.cursor()
cursor.execute(q8)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

```python
q9 = """
SELECT
    CASE
        WHEN MONTH(Case_Date) IN (12, 1, 2) THEN 'Winter'
        WHEN MONTH(Case_Date) IN (3, 4, 5) THEN 'Spring'
        WHEN MONTH(Case_Date) IN (6, 7, 8) THEN 'Summer'
        ELSE 'Autumn'
    END AS season,
    COUNT(*) AS total_crimes,
    SUM(CASE WHEN Domestic = 1 THEN 1 ELSE 0 END) AS domestic_crimes,
    SUM(CASE WHEN Domestic = 0 THEN 1 ELSE 0 END) AS non_domestic_crimes
FROM CCA.crime_data
GROUP BY season
ORDER BY FIELD(season, 'Winter', 'Spring', 'Summer', 'Autumn');
"""

cursor = connection.cursor()
cursor.execute(q9)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```

```python
q10 = """
SELECT
Block,
DATE_FORMAT(Case_Date, '%Y-%m') AS Crime_Month,
COUNT(*) AS total_crimes
FROM CCA.crime_data
GROUP BY Block, Crime_Month
ORDER BY total_crimes DESC
LIMIT 20;
"""
cursor = connection.cursor()
cursor.execute(q10)
out = cursor.fetchall()
print(tabulate(out, headers=[i[0] for i in cursor.description], tablefmt='psql'))
```
