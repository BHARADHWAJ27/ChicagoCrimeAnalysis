```python
import pandas as pd
import numpy as np
import zipfile
```

```python
data=pd.read_excel(r"/content/Crime_Data.xlsx")
display(data.head())
```

```python
data.info()
```

```python
data['Location Description'].fillna('Unknown',inplace=True)
```

```python
data.isnull().sum()
```

```python
blockcoordinates=data.groupby('Block')[['Latitude','Longitude']].first()
data['Latitude']=data['Latitude'].fillna(data['Block'].map(blockcoordinates['Latitude']))
data['Longitude']=data['Longitude'].fillna(data['Block'].map(blockcoordinates['Longitude']))
```

```python
data.drop(['Ward','Community Area'],axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Longitude']=data['Longitude'].fillna('Unknown')
data['Latitude']=data['Latitude'].fillna('Unknown')


```

```python
data[pd.to_numeric(data['Latitude'], errors='coerce').notnull()]
data['Latitude'] = pd.to_numeric(data['Latitude'], errors='coerce')
data['Latitude'] = pd.to_numeric(data['Latitude'], errors='coerce').fillna(0.0)

data[pd.to_numeric(data['Longitude'], errors='coerce').notnull()]
data['Longitude'] = pd.to_numeric(data['Longitude'], errors='coerce')
data['Longitude'] = pd.to_numeric(data['Longitude'], errors='coerce').fillna(0.0)
```

```python
data.isnull().sum()
```

```python
data['Case_Date'] = data['Date'].dt.date
data['Time'] = data['Date'].dt.time
```

```python
print(data[['Date', 'Time']].head())
```

```python
data.drop('Date',axis=1,inplace=True)
```

```python
data['Location'] = data.apply(
    lambda row: f"({row['Latitude']}, {row['Longitude']})" if not pd.isna(row['Latitude']) and not pd.isna(row['Longitude']) else None,
    axis=1
)
data.head()
```

```python
data['Update_Date'] = data['Updated On'].dt.date  # Extract only the date part
data['Update_Time'] = data['Updated On'].dt.time  # Extract only the time part

# Verify the result
print(data[['Updated On', 'Update_Date', 'Update_Time']].head())
```

```python
data.isnull().sum()
```

```python
data.drop(['X Coordinate','Y Coordinate'],axis=1,inplace=True)
```

```python
data['IUCR'] = pd.to_numeric(data['IUCR'], errors='coerce')
data = data.dropna(subset=['IUCR'])  # Removes rows where 'iucr' couldn't be converted
data['IUCR'] = data['IUCR'].astype('int64')
```

```python
data.rename(columns={'Location Description': 'Location_Description'}, inplace=True)
data.rename(columns={'Primary Type':'Primary_Type'},inplace=True)
data.rename(columns={'Case Number':'Case_Number'},inplace=True)
#data.rename(columns={'Location':'Location'},inplace=True)

# Save back to CSV (optional)
#df.to_csv('your_file_cleaned.csv', index=False)
```

```python
#data.drop('Updated On',axis=1,inplace=True)
data['Primary_Type']=data['Primary_Type'].astype('category')
data['Location_Description']=data['Location_Description'].astype('category')
```
